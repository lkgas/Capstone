```python
len(scores['Epoch'])
```

```python
epoch = np.array(scores['Epoch'])
train_MAPE = np.array(scores['Train_MAPE'])*100
train_RMSE = np.array(scores['Train_RMSE'])
test_MAPE = np.array(scores['Test_MAPE'])*100
test_RMSE = np.array(scores['Test_RMSE'])
```

```python
idx1 = np.array(test_MAPE).argmin()
idx2 = np.array(test_RMSE).argmin()
```

```python
fig, ax1 = plt.subplots()

ax1.grid(b=True, color='gray', linestyle='--', linewidth=1, alpha=0.3)
text = "Epoch={:d}, Min Test MAPE={:.4f}%".format(epoch[idx1], test_MAPE[idx1])

ax1.plot(epoch, train_MAPE, label='Train MAPE', color='r', linewidth=2, linestyle='-')
ax1.plot(epoch, test_MAPE, label='Test MAPE', color='b', linewidth=2, linestyle='-')

ax1.annotate(text, xy=(epoch[idx1], test_MAPE[idx1]))

ax1.set_xlabel('Epoch')
ax1.set_ylabel('MAPE (%)')
ax1.legend()
```

```python
fig, ax1 = plt.subplots()

ax1.grid(b=True, color='gray', linestyle='--', linewidth=1, alpha=0.3)
text = "Epoch={:d}, Min Test RMSE={:.6f}".format(epoch[idx1], test_RMSE[idx1])

ax1.plot(epoch, train_RMSE, label='Train RMSE', color='r', linewidth=2, linestyle='-')
ax1.plot(epoch, test_RMSE, label='Test RMSE', color='b', linewidth=2, linestyle='-')

ax1.annotate(text, xy=(epoch[idx1], test_RMSE[idx1]))

ax1.set_xlabel('Epoch')
ax1.set_ylabel('RMSE')
ax1.legend()
```
